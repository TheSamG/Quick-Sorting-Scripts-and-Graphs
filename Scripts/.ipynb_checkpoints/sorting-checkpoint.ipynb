```python
import os
import numpy as np
import pandas as pd
```

```python
year = "2020"
```

```python
output_path = r"C:\Users\samgh\Downloads\Code\Jacob Files\Output"
output_path = os.path.join(output_path, year + " aggregated.csv")
```

```python
csv_dir = r"C:\Users\samgh\Downloads\Code\Jacob Files\CSVs"
csv_path = os.path.join(csv_dir, year + " raw data.csv")
```

```python
physicians = pd.read_csv(csv_path) # load to DataFrame
physicians
```

```python
specialty = physicians.groupby(['physician_profile_id']).max()
a = specialty[['physician_specialty']]
a
```

```python
money = physicians.groupby(['physician_profile_id']).sum()
b = money[["total_amount_of_payment_usdollars"]]
b
```

```python
final = a.join(b).reset_index()
final
```

```python
final.to_csv(output_path)
```
